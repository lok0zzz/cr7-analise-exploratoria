**LIMPEZA DOS DADOS**

```python
df = pd.read_csv('data.csv')
del df['Unnamed: 0']
df
```

```python
display(df.isna().sum())
print('-------------------------------')
```

```python
agrupado = df.Position
print(agrupado.value_counts())
df.Position.fillna('LW', inplace=True)
df
```

```python
agrupado = df.Type_of_goal
print(agrupado.value_counts())
df.Type_of_goal.fillna('Right-footed shot', inplace=True)
df
```
